# Face Detection Model

## 1.1 Collect images using opencv 

```python
import os
import time
import uuid # uniform unique identifier gives unique identities to images instead of labeling them as img1,img2 etc
import cv2
```

```python
image_path = os.path.join('data','images')
number_images=30
```

```python
cap = cv2.VideoCapture(0)
for imgnum in range(number_images):
    print('Collecting image {}'.format(imgnum))
    ret, frame = cap.read()
    imgname = os.path.join(image_path,f'{str(uuid.uuid1())}.jpg')
    cv2.imwrite(imgname, frame)
    cv2.imshow('frame', frame)
    time.sleep(0.5)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

### 2.3 Load image into TF data pipeline

```python
images = tf.data.Dataset.list_files('/Users/abhigyantripathi/Desktop/PROJECTS/data/images/*.jpg') 
## search for any file with .jpg extension
# print("Current working directory:", os.getcwd())
```

```python
images.as_numpy_iterator().next() ## checking if images has picked up any data
```

```python
def load_image(x): 
    byte_img = tf.io.read_file(x)
    img = tf.io.decode_jpeg(byte_img)
    return img
```

```python
images = images.map(load_image) ## the map function copies the load_image function to each member in the images path
```

```python
images.as_numpy_iterator().next()
```

```python
type(images)
```

### 2.4 View raw images in matplotlib

```python
image_generator = images.batch(4).as_numpy_iterator() 
# creates a batch of 4 images
```

```python
plot_images = image_generator.next()
# each time you run it will give batches of 4
```

```python
fig, ax = plt.subplots(ncols=4, figsize=(20,20))
for idx, image in enumerate(plot_images):
    ax[idx].imshow(image) 
plt.show()
# function to convert images to their visible format
```

### 3.2 Move the matching labels

```python
# print("Current working directory:", os.getcwd())
for folder in ['train','test','val']:
    for file in os.listdir(os.path.join('/Users/abhigyantripathi/Desktop/PROJECTS/data', folder, 'images')):
        
        filename = file.split('.')[0]+'.json'
        existing_filepath = os.path.join('/Users/abhigyantripathi/Desktop/PROJECTS/data','labels', filename)
        if os.path.exists(existing_filepath): 
            new_filepath = os.path.join('/Users/abhigyantripathi/Desktop/PROJECTS/data',folder,'labels',filename)
            os.replace(existing_filepath, new_filepath)      
```
